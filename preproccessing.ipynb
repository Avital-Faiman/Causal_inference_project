### Adding T variable to the dataset


```python
threshold = 8
math_data['T'] = ((math_data['absences']) >= threshold).astype(int)
por_data['T'] = ((por_data['absences']) >= threshold).astype(int)

math_data['Y'] = (math_data['G1'] + math_data['G2'] + math_data['G3']) / 3
por_data['Y'] = (por_data['G1'] + por_data['G2'] + por_data['G3']) / 3
```

## Outcome
Finding the best outcome for our data --> the absences are given as a number so we want to find the best threshold.
We tried to use several methods to find the best threshold.

```python
# Define the threshold as the 75th percentile of the 'absences' column
threshold_math = np.percentile(math_data['absences'], 75)
threshold_por = np.percentile(por_data['absences'], 75)

# Create a new column 'absences_lots' that is 1 if 'absences' is above the threshold, 0 otherwise
math_data1 = math_data
por_data1 = por_data
math_data1['absences_lots'] = (math_data1['absences'] > threshold_math).astype(int)
por_data1['absences_lots'] = (por_data1['absences'] > threshold_por).astype(int)

# Print the number and percentage of students with a lot of absences
num_lots_math = math_data1['absences_lots'].sum()
num_lots_por = por_data1['absences_lots'].sum()

pct_lots_math = num_lots_math / len(math_data) * 100
pct_lots_por = num_lots_por / len(math_data) * 100
```

## Merging of math and por datasets into one
There isn't id that can be used to identify students, so in order to find students that take both courses we will group them with the following features that we assume can identify the
students : "school", "sex", "age", "address", "famsize", "Pstatus", "Medu", "Fedu", "Mjob", "Fjob", "reason", "internet".
We are also adding random sampling for students that take math and por.

```python
#We want to find the number of students that take both math and por
udf = pd.concat([math_data, por_data])

q1 = """select school, sex, age, address,famsize,  Pstatus, Medu, Fedu, Mjob, Fjob, reason, internet
        from  udf
        group by school, sex, age, address,famsize,  Pstatus, Medu, Fedu, Mjob, Fjob, reason, internet
        having count(*) > 1"""

print(ps.sqldf(q1, locals()))
```

```python
#we want to check the correlation between the datasets, we will find pearson correlation between the grades of por and the grads of math.
math_data2 = math_data
por_data2 = por_data

#adding another col to separate math and por
math_data2["type"] = 1
por_data2["type"] = 0

u = pd.merge(math_data2, por_data2, on = ["school", "sex", "age", "address", "famsize", "Pstatus", "Medu", "Fedu", "Mjob", "Fjob", "reason", "internet"])

pearson_correlation = u['avg_3_y'].corr(u['avg_3_x'])
print("Pearson correlation coefficient:", pearson_correlation)
```
